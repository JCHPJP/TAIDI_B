```python
title  = []
import os
folder_path='processed'
for root, dirs, files in os.walk(folder_path):
        for file in files:
            if file.endswith('.md'):
                # md_files.append(os.path.join(root, file))
                with open(os.path.join(root, file), 'r', encoding='utf-8') as f:
                    text = f.read()

# 按 # 分割（会丢失 # 符号）
                    sections = text.split('#')
                    for i in sections:
                        title.append(i[:i.find("\n")])
```

```python
for i in set(title):print(i)
```

如果段落包含保留关键字 → 保留

如果段落包含删除关键字 → 删除

其他段落 → 默认保留
